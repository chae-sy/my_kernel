```python
import torch

# 행렬 크기
m, n, k = 16, 8, 16

# hA 초기화 (row-major)
hA = torch.zeros((m, k), dtype=torch.float16)
for r in range(m):
    for c in range(k):
        hA[r, c] = ((r + c) % 7) * 0.1

# hB 초기화 (column-major)
# 실제 메모리 배치는 row-major지만, 값은 column-major 패턴에 맞게 채움
hB = torch.zeros((k, n), dtype=torch.float16)
for c in range(n):
    for r in range(k):
        hB[r, c] = ((r + c) % 5) * 0.1

print("hA (row-major):")
print(hA)

print("\nhB (column-major 패턴으로 채워짐):")
print(hB)

# 행렬곱 결과
C = torch.matmul(hA, hB)
print("\nResult (C = A × B):")
print(C)
```
